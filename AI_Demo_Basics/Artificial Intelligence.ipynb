```python
x =Data['SAT']
y =Data['GPA']
```

```python
plt.scatter(x,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
z= x+100
k= y+50
```

```python
plt.scatter(z,k)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
x_1 = sm.add_constant(x)
x_1
```

```python
model = sm.OLS(y, x_1).fit()
model.summary()
```

```python
plt.scatter(x,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
m, c = np.polyfit(x,y,1)
n_y = m*x +c
plt.plot(x,n_y)
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
m, c = np.polyfit(x,y,1)
n_y = m*x +c
plt.plot(x,n_y, color ='red')
plt.show()
```
